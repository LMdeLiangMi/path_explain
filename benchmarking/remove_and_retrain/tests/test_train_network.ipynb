```python
x = np.random.randn(2000, 10).astype(np.float32)
y = 2 * x[:, 0] * x[:, 1] - \
    x[:, 0] * x[:, 2] + \
    1.5 * x[:, 2] * x[:, 3] + \
    0.5 * x[:, 1] * x[:, 3] + \
    3 * x[:, 6] * x[:, 7] - \
    2 * x[:, 8] * x[:, 9] + \
    x[:, 4] - x[:, 5]
x_train, y_train = x[:400], y[:400]
x_test, y_test = x[400:], y[400:]
```

```python
model = interaction_model(num_features=x_train.shape[1],
                          num_layers=2,
                          hidden_layer_size=4,
                          interactions_to_ignore=None,
                          regression=True)
compile_model(model,
              learning_rate=0.15,
              regression=True)
model.fit(x=x_train,
          y=y_train,
          batch_size=128,
          epochs=100,
          verbose=1)
model.evaluate(x_test, y_test)
```
